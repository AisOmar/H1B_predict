### Random Forest 

```python
from sklearn import datasets, model_selection
from sklearn import linear_model, metrics
from sklearn import preprocessing, pipeline
from sklearn.metrics import classification_report
from sklearn.svm import LinearSVC, SVC
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.metrics import accuracy_score, fbeta_score, f1_score, precision_score, recall_score
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
X = df.drop('CASE_STATUS',axis=1)
y = df.CASE_STATUS
```

```python
X_train, X_test, y_train,y_test = model_selection.train_test_split(X,y, test_size=0.2)
```

```python
brf = BalancedRandomForestClassifier(max_depth=40,n_estimators=100, random_state=0)
```

```python
brf.fit(X_train, y_train)
y_pred_r = brf.predict(X_test)
print(classification_report(y_test, y_pred_r))
```

```python
print(metrics.confusion_matrix(y_test,y_pred_r))
```

```python
logit_roc_auc = roc_auc_score(y_test, y_pred_r)
```

```python
print(logit_roc_auc)
```
